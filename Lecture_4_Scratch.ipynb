```python
from thinkdsp import CosSignal, SinSignal
from thinkdsp import decorate
```

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
cos1250 = CosSignal(freq=1250, amp=1, offset = 0)
cos1250.plot()
```

```python
segment = cos1250.make_wave(3, framerate = 1000)
```

```python
print(len(segment.ys))
segment.plot()
```

```python
spectrum = segment.make_spectrum()
spectrum.plot()
```

```python
sin_sig = SinSignal(freq=1, amp=1, offset=0)
sin_sig2 = SinSignal(freq=8, amp=0.2, offset=0)
mix = sin_sig + sin_sig2
mix.plot()
decorate(xlabel='Time (s)')
```

```python
segment2 = mix.make_wave(duration=3, start = 0, framerate = 100)
```

```python
spectrum2 = segment2.make_spectrum()
```

```python
spectrum2.plot(high=10)
```
